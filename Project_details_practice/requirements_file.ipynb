## -e. = Install the current project as a package in editable (development) mode

#### An editable install (using -e .) in Python means that your project is linked, not copied, into your environment. As you develop your code (for example, a custom machine learning package), any changes you make in your source files are immediately reflected when you import and run your code. For example, if you’re building a library to preprocess data, you can keep editing your functions, and the next time you run your script, your updated code is used—no need for reinstallation.

#### 1️. What does (.) mean?
    -> means current directory
    -> So this refers to your project folder, the one that contains setup.py

#### 2. What does (-e) mean?
    -> -e stands for editable
    -> Editable = changes in your source code are immediately reflected
    -> You don’t need to reinstall every time you change code

#### 3. What happens when pip reads (-e .)?
    -> Looks at the current folder
    -> Finds setup.py (or pyproject.toml)
    -> Runs the setup process
    -> Links your project into the Python environment
    -> Does not copy your source code
    -> Registers your project as an importable package
