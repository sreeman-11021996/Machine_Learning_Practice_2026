### SYS Libraries = is a built-in standard library that provides access to variables, functions, and objects that interact closely with the Python interpreter and the runtime environment.

### exc_info() = exc_type, exc_value, exc_traceback
### Def : exc_info() is a function from the sys module that tells you:
    “What exception just happened right now?”
    It is mainly used inside an except block.

```python
# Syntax
import sys
sys.exc_info()
```

```python
# exc_type
"""The class/type of the exception."""
"""Ex. ValueError, ZeroDivisionError, TypeError"""

try:
    1 / 0
except:
    exc_type, exc_value, exc_tb = sys.exc_info()
    print(exc_type)
```

```python
# exc_value
"""The exception object itself, including the error message."""

print(exc_value)
```

```python
# exc_traceback
"""This is the traceback object — it contains the entire call stack of where the error occurred."""

"""Every exception object has an attribute:
    exc_traceback = exception.__traceback__ = exc_value.__traceback__"""

try:
    1 / 0
except Exception as e:
    print(e.__traceback__)
```

```python
# exc_traceback : details of object
"""A traceback is a linked list of stack frames.
    Each traceback node contains:

    tb_frame → the stack frame
    tb_lineno → line number
    tb_next → next traceback (previous call)"""
```

```python
# tb_frame : “One function call at one moment in time”
"""Contains :
    -> Local Variables
    -> Global Variables
    -> Code Object : f_code
    -> Filename"""

exc_traceback = exc_tb
frame = exc_traceback.tb_frame
print(frame)
```

```python
# f_code : is a code object describing the function being executed.
"""Contains :
    -> co_name = Function name
    -> co_filename = file path
    -> co_firstlineno = where function starts"""

code = frame.f_code
print(code.co_name)
print(code.co_filename)
print(code.co_firstlineno)
```

## Implementing the above together :

```python
import sys
```

```python
def divide():
    return 1 / 0
```

```python
try:
    divide()
except Exception as e:
    exc_type, exc_value, exc_tb = e.exc_info()

    print("Type:", exc_type)
    print("Message:", exc_value)

    frame = exc_tb.tb_frame
    code = frame.f_code

    print("Function:", code.co_name)
    print("File:", code.co_filename)
    print("Line:", exc_tb.tb_lineno)
```

1. exc_info() = type(e), e, e.__traceback__

This function returns the old-style representation of the handled exception. If an exception e is currently handled (so exception() would return e), exc_info() returns the tuple (type(e), e, e.__traceback__). That is, a tuple containing the type of the exception (a subclass of BaseException), the exception itself, and a traceback object which typically encapsulates the call stack at the point where the exception last occurred.
