```python
import numpy as np
import pandas as pd
```

#### 2. Create a Dataframe

```python
# pd.DataFrame (data : ndarray, index : list, columns : list , copy)
# copy = True (new memory allocated, Forces Pandas to not reuse underlying NumPy memory), 
#        False (shares memory with original)
df = pd.DataFrame(data= np.arange(0,20).reshape(5,4), 
                    index=['Row1', 'Row2','Row3', 'Row4','Row5'],
                    columns= ['Column1','Column2','Column3','Column4']
                   )
```

#### 3. Get Details on Data Frame

```python
# get the first five rows
df.head()
```

```python
# save in a csv file (coma seperated values)
df.to_csv('Pandas_Test_1.csv')
```

#### 3. Accessing Elements

##### A. Direct 

```python
df['Column3']
```

```python
type(df['Column3'])
```

```python
# df[list of column names]
df[['Column1','Column2']]
```

##### B. df.loc [row, col] : data location (focus on row indexes mainly)

```python
## Series o/p
df.loc['Row1']
```

```python
## Series o/p
df.loc[:,'Column1']
```

```python
df.loc[df['Column1']<10, ['Column2','Column1']]
```

##### C. df.iloc [row, col]  : (mainly focuses on columns and rows)

```python
# : all rows / columns
df.iloc[:,:]
```

```python
# dataframe
df.iloc[0:2,0:2]
```

```python
# series
df.iloc[0:2,0]
```

```python
# remove the elements of column1
df.iloc[:,1:]
```
