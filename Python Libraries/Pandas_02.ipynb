#### 3. CSV Files reading : parameters


```python
# dummy csv file
data_1 = ('col1,col2,col3\n'
            'x,y,1\n'
            'a,b,2\n'
            'c,d,3')

data_2 = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11,12')

data_3 = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')

data_4 = ('a,b,c\n'
           '4,apple,bat\n'
            '8,orange,cow')

data_5 = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

##### A. (sep : 'str')

```python
# read from StringIO
pd.read_csv(StringIO(data_1), sep=',')
```

##### B. (usecols : list)

```python
# usecols : read from specific columns
pd.read_csv(StringIO(data_1), usecols=['col1', 'col3'])
```

##### C. (dtype : dict / the data type)

```python
# dtype : specify data types for the columns
df_sample_1 = pd.read_csv(StringIO(data_2), dtype = float)
df_sample_1
```

```python
# dtype : specify different data types for columns
df_sample_2 = pd.read_csv(StringIO(data_2), dtype = {'a':int, 'b':object, 'c':float, 'd':'Int64'})
df_sample_2
```

```python
def get_column_dtypes (df:pd.DataFrame):
    column_dtype_dict = {}
    for col in df.columns:
        column_dtype_dict[col] = df[col].dtype.name
    return column_dtype_dict
```

```python
get_column_dtypes(df_sample_2)
```

##### D. (index_cols : int / Boolean)

```python
df_sample_3 = pd.read_csv(StringIO(data_3), index_col=0)
df_sample_3
```

```python
df_sample_3 = pd.read_csv(StringIO(data_3), index_col=1)
df_sample_3
```

```python
df_sample_4 = pd.read_csv(StringIO(data_4), index_col=False)
df_sample_4
```

##### E. (escapechar : str)

```python
df_sample_5 = pd.read_csv(StringIO(data_5), escapechar='\\')
df_sample_5
```
